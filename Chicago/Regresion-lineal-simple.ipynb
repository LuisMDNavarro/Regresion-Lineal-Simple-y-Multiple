```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

**Private_room**

```python
#Cargar un archivo csv
df = pd.read_csv('Private_room.csv')

# Lista de variables independientes
vars_independientes = ['host_acceptance_rate', 'host_is_superhost', 'bathrooms', 
                       'review_scores_cleanliness', 'host_identity_verified', 'instant_bookable',
                       'property_type']
# Lista de variables dependiente
vars_dependientes = ['price', 'price', 'accommodates', 'price', 'price', 'price', 'price']
# DataFrame vacío para almacenar los coeficientes de correlacion R
heatmap_data = pd.DataFrame(index=vars_dependientes, columns=vars_independientes)
resultados = []
mayor_r = 0
mejor_modelo = {}

for var_indep, var_dep in zip(vars_independientes, vars_dependientes):
    Vars_Indep = df[[var_indep]] 
    Var_Dep = df[var_dep]
    #Se define model como la funcion de regresion lineal
    model = LinearRegression()
    #Ajustamos el modelo con las variables antes declaradas
    model.fit(X = Vars_Indep, y = Var_Dep)
    #Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R Determinacion
    coef_Deter = model.score(X = Vars_Indep, y = Var_Dep)
    #Corroboramos cual es el coeficiente de Correlacion de nuestro modelo
    coef_Correl = np.sqrt(coef_Deter)
    # Guardar el valor en la matriz del heatmap
    heatmap_data.loc[var_dep, var_indep] = coef_Correl

    # Guardar en la lista de resultados
    resultados.append({
        "Variable dependiente": var_dep,
        "Variable independiente": var_indep,
        "R": coef_Correl,
        "R2": coef_Deter
    })

    # Si es la mejor correlación hasta ahora, guardamos el modelo
    if coef_Correl > mayor_r:
        mayor_r = coef_Correl
        mejor_modelo = {
            "var_dep": var_dep,
            "var_indep": var_indep,
            "a": model.coef_[0],
            "b": model.intercept_,
            "R": coef_Correl,
            "R2": coef_Deter
        }

# Convertir resultados a DataFrame y ordenar por R descendente
df_resultados = pd.DataFrame(resultados).sort_values(by="R", ascending=False)
#Eliminar filas repetidas
heatmap_data = heatmap_data[~heatmap_data.index.duplicated(keep='first')]
# Convertir valores a tipo numérico (pueden ser strings por NaN)
heatmap_data = heatmap_data.astype(float)

#Mapa de calor
plt.figure(figsize=(10, 2.5))
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", fmt=".4f", linewidths=0.5)
plt.title("Coeficiente de Correlación (R)")
plt.xlabel("Variables Independientes")
plt.ylabel("Variables Dependientes")
plt.show()

# Tabla de resultados
print("Tabla de Resultados (R y R^2)")
print(df_resultados.to_string(index=False))
print("\n\n")

#Modelo matematico
print(mejor_modelo['var_dep'], " vs ", mejor_modelo['var_indep'])
print(f"Modelo matemático: y = {mejor_modelo['a']:.4f}x + {mejor_modelo['b']:.4f}")
print(f"Correlación R: {mejor_modelo['R']:.4f} | Determinación R^2: {mejor_modelo['R2']:.4f}")
```
