```python
import pandas as pd
df = pd.read_csv('50_sin_nulos_ni_atipicos_Chicago_Illinois_UnitedStates.csv')
df = df.drop(['Unnamed: 0'], axis = 1)
df.head(5)
```

```python
#Cambiar boleano a numerico en  host_identity_verified, host_is_superhost y instant_bookable
df["host_identity_verified"] = df["host_identity_verified"].replace({"t": 1, "f": 0})
df["host_is_superhost"] = df["host_is_superhost"].replace({"t": 1, "f": 0})
df["host_is_superhost"] = df["host_is_superhost"].replace({"Unknown": 0})
df["instant_bookable"] = df["instant_bookable"].replace({"t": 1, "f": 0})
#Cambio a numerico por jerarquia de property_type y room_type
    # Contar frecuencia de cada tipo de propiedad
frecuencia = df['property_type'].value_counts()
frecuencia2 = df['room_type'].value_counts()
    # Crear un diccionario con los valores numÃ©ricos segÃºn la frecuencia
property_map = {tipo: idx for idx, tipo in enumerate(frecuencia.index)}
property_map2 = {tipo: idx for idx, tipo in enumerate(frecuencia2.index)}
    # Reemplazar en el DataFrame
df['property_type'] = df['property_type'].map(property_map)
df['room_type'] = df['room_type'].map(property_map2)
df.head(5)
```

```python
#['Private room' 'Entire home/apt' 'Shared room' 'Hotel room']
filtro1 = df[df["room_type"] == 1]
filtro2 = df[df["room_type"] == 0]
filtro3 = df[df["room_type"] == 3]
filtro4 = df[df["room_type"] == 2]
```

```python
filtro1.to_csv("Private_room.csv", index=True)
filtro2.to_csv("Entire_home_apt.csv", index=True)
filtro3.to_csv("Shared_room.csv", index=True)
filtro4.to_csv("Hotel_room.csv", index=True)
```
