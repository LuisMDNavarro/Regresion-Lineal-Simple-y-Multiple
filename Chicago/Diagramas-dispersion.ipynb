```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

**Private_room**

```python
#Cargar un archivo csv
df = pd.read_csv('Private_room.csv')

# Lista de variables independientes
vars_independientes = ['host_acceptance_rate', 'host_is_superhost', 'bathrooms', 
                       'review_scores_cleanliness', 'host_identity_verified', 'instant_bookable',
                       'property_type']
# Lista de variables dependiente
vars_dependientes = ['price', 'price', 'accommodates', 'price', 'price', 'price', 'price']

colores = ["red", "blue", "green", "purple", "orange", "black", "magenta"]
for var_indep, var_dep, color in zip(vars_independientes, vars_dependientes, colores):
    sns.scatterplot(x = var_indep,  y = var_dep, color = color, data = df)
    plt.title(f"{var_dep} vs {var_indep}")
    plt.show()
```
